The color palette reveals a lot about the electronic design of the device or technology, and its visual capabilities for human users.

in this notebook I will show you a method to find the color palette of a image

### Import packages

```python
from PIL import Image,ImageShow, ImageDraw 
from IPython.display import display
import matplotlib.pyplot as plt
import extcolors
```

### Import image from path

<b>- Define the image path</b>

```python
path = "images/image.jpg"
```

<b>- Open the image in a variable "img"</b>

```python
try:  
    img = Image.open(path)  
except IOError: 
    pass
display(img.resize((400,400)))
```

# 

## Extract the colors and pixel count

#### - "Colors" variable is a list that contain the RGB color and the pixels counts for each color

```python
colors, pixel_count = extcolors.extract_from_image(img)
```

#### - Show the colors in RGB format, with the pixels count of each color

```python
print("Pixel Count: ",pixel_count)
print(" ")
print("Colors:")
for color in colors:
    print(color)
```

#  

<h2>Print colors</h2>

#### - Create the squares shape

```python
w, h = 100, 100
shape = [(0, 0), (w, h)] 
```

#### - Show the color squares

```python
for i in range(0,len(colors)):
    
    # Calculate the % of the image whit this color
    percent = ((colors[i][1]/pixel_count)*100)
    
    # % Limit
    if percent > 0.05:
        uno = Image.new("RGB", (w, h)) 

        # Create rectangle image 
        dos = ImageDraw.Draw(uno)   
        dos.rectangle(shape, fill =colors[i][0]) 
        
        # Show
        display(uno,("%.2f" % percent)+"%") 
```
